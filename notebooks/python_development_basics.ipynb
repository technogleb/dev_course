## О языке Python.

#### PEP8

В коммьюнити питона каждые предложения по изменению в язык оформляются в виде специального документа Python Enhancement Proposal, сокращенно PEP. Далее его рассматривают и принимают, либо не принимают.  
Один из самых первых и известных PEP-ов это PEP8 - документ, описывающий то, как должен выглядеть код на python.
Как назыавть переменные, какие делать отсупы, сколько должно быть аргументов у методов, и так далее. Документ очень большой и полностью может быть прочитан по ссылке - https://www.python.org/dev/peps/pep-0008/.

Здесь мы приведем основные моменты.
* используем snake_case для именования переменных
* исключение составляют классы - их пишем CamelCase-ом
* отступы - четыре пробела
* две пустых строки между функциями, классами, одна между методам класса
* длина строки 80 или 120 символов (в зависимости от вашего стайлгайда)
* и многое другое

Все, что не оговорено в PEP8, остается на усмотрение вашей команды. Главное, это выработать собственный набор правил и соблюдать его везде.

## Pip, cторонние библиотеки и виртуальное окружение.

#### Пакетный менеджер pip

Не смотря на довольно богатую стандартную библиотеку, большинство вещей, которые в будете делать в питоне скорее
всего будут использовать код, написанный кем-то другим. В питоне, как и во многих других языках есть свой пакетный
менеджер - программа, позволяющая устанавливать, обновлять и удалять сторонние пакеты. В питоне такой пакетный менеджер называется PIP. 

PIP является в курсе всех пакетов, загружаемых в специальный репозиторий для публичный модулей - PyPI (python package index). Каждый пакет вы можете добавить к себе в область видимости вашего питона, используя консольный интерфейс PIP-a.

Например, данная команда установит пакет с именем numpy.

```python
pip install numpy
```

Также можно указать конкретную версию желаемого пакета 

```python
pip install numpy==1.16.3
```

Данная команда обновит numpy до последней доступной и совместимой версии.

```python
pip install -U numpy
```

Данная команда удалит numpy.

```python
pip uninstall numpy
```

Данная команда выведет все пакеты в текущем окружении и их версии.

```python
pip list
```

#### Виртуальные окружения

По умолчанию в вашей системе наверняка есть python, в большинстве случаев даже python3. Однако находится он там не для вас, а потому что сама система его использует (т.е. некоторые верхнеуровневые части ОС написаны на питоне). У каждого такого pythona есть свое окружение и свои библиотеки. Проблемы с использованием стандартного питона заключается в том, что вы можете что-нибудь там поломать, тогда ваша система может работать некорректно. Кроме того, с каждым новым проектом вы будете все больше и больше захламлять системное окружение. Более того, если вы захотите работать над двумя проектами с двумя разными окружениями, то у вас не получится этого сделать ввиду очевидных причин. Отсюда вытекает must-have практика в любой разработке на python - использование виртуального окружения.

По-умолчанию, все пакеты которые вы ставите в через pip, ставятся в так называемое системной окружение, то есть окружение по умолчанию. Однако, в ситуации, когда у вас есть несколько проектов, каждый из которых должен использовать разные версии питона и различных пакетов, происходит конфликт интересов, из которого следует необходимость разграничиавть зону ответственности. Такая "зона отвественности" и называется окружение. 

В питоне существует несколько инструментов, используемых для создания окружения. Мы будем использовать два из них, как наиболее распространенные и официально поддерживаемые virtualenv и venv.

Virtualenv - это классический инструмент, который устанавливается командой "pip install virtualenv" и совместим как со вторым питоном так и с третьим.

venv - официальная утилита, которая идет по умолчанию вместе с третьим питоном. Если вы точно знаете, что будете работать только с 3-им питоном, рекомендуется использовать его.

#### Как установить виртуальное окружение?

* с помощью virtualenv  
virtualenv -p <python_version/path> env_name  
* c помощью venv  
<python_version/path> -m venv env_name

В примерах выше <python_version/path> это либо версия вашего питона, с которым вы хотите установить виртуальное окружение, в случае если нужная версия в переменной PATH, либо непосредественно путь до бинарника вашего питона.  
  
Так, команда <<python3.6 -m venv env>> установит окружение с именем env и версией питона 3.6. Если бы ваш терминал не видел python3.6, можно было бы напрямую прописать путь до нужного питона - <</usr/bin/python3.6 -m venv env>>

Если у вас нет необходимой версии питона, ее можно установить. Это рекомендуется делать утилитой pyenv.

Часто возникает вопрос про использование проекта anaconda. Anaconda неофицальный пакетный менеджер (а также менеджер окружений) для научноориентированного питона.
Официальным пакетным менеджером для питона является pip! Рекомендуется использовать его!
Зачем же тогда нужна conda? Conda нужна когда есть существенные требования к скорости работы таких пакетов как scipy, numpy. Conda была написана таким образом, чтобы соптимизировать их работу на большинстве процессоров и ускорить до 30%.

## Модули и система импорта в питоне

Разберем очень важную тему, которую почему-то пропускают в большинстве курсов по питону - порядок импорта в питоне.
За него отвечает переменная sys.path. Вообще, sys ваш лучший друг.

```python
import sys
sys.path
```

Рассмотрим, из чего состоит sys.path. sys.path - это список путей, в которые смотрит питон, когда вы что-либо импортируете.  
Причем смотрит он в порядке расположения строк в списке.  
  
В первую очередь в sys.path автоматически добавляется текущая директория, из который исполняется скрипт - в нашем случае /Users/gsinyakov/ts_course/
Поэтому, если вы создадите в текущей директории файлик numpy.py и попытаетесь импортировать что-либо из нумпая, вы будете импортировать именно этот файлик, а не библиотеку numpy, которую ожидаете. 
  
Далее по списку следуют системные строки, которые ссылаются на стандартную библиотеку питона и всегда добавляются в sys.path по умолчанию. Именно поэтому вас всегда доступны такие штуки как math, random, sys, и так далее.  
  
Далее следуют так называемые site-packages - это такая папка, куда устанавливаются все библиотеки командой pip install или python setup.py install. Все сторонние библиотеки всегда идут в эту папку.
По умолчанию site-packages заканчивают список, но в нашем случае туда еще попали пути, которые хочет сохранить jupyter для своих нужд.
  
На самом деле, вы можете добавить произвольные пути в sys.path, которые вам захочется. Можно это сделать при помощи команды sys.path.append('pathfile'), однако для этого существует специальная переменная окружения, которую задают из bash'a и которая называется PYTHONPATH. Все, что перечислено в этой переменной, автоматически добавляется в sys.path между текущей директорией и путями системной библиотеки. Например, в нашем случае, перед тем как запустить jupyter notebook, я прописал в том же терминале export PYTHONPATH=$PYTHONPATH:/Users/gsinyakov/ts_course. Ее, собственно, можно наблюдать второй строкой в sys.path.
  
Не путать ее с первой строкой /Users/gsinyakov/ts_course ! Первая строка означает текущую директорию, просто в данном случае они совпали. Если мы перейдем из текущей директории, первая строка изменится, вторая нет.
Теперь вопрос - зачем я добавил в sys.path путь /Users/gsinyakov/ts_course? Это было сделано для того, чтобы я всегда мог писать абсолютные импорты относительно корня проекта и не волноваться о том, из какой директории я сейчас запускаю скрипт/юпитер.
  
*примечание - единственное, что не входит в sys.path - это билтины built-ins, такие как list, int, float и так далее. Они являются часть низкоуровнего кода питона и всегда доступны для импорта. Переопределить их из текущей директории не получится :)

  
##### Итак, рекомендация. Устанавливать PYTHONPATH в корень вашего проект, чтобы всегда можно было использовать свой код.

  
Примечание для продвинутых*
Проблема c PYTHONPATH заключается в том, что эту строчку придется каждый раз прописывать в каждом новом терминале, откуда вы собираетесь запускать код (юпитер в том числе). Если вы хотите установить ваш код перманентно в текущее окружение, вам необходимо будет написать скрипт setup.py. Подробнее можете ознакомиться здесь - https://pythonhosted.org/an_example_pypi_project/setuptools.html

## Немного о GIT.

GIT - самая распространенная система контроля версий в мире. Система контроля версий - это такая программа, которая позволяет сохранять все изменения нашего кода, которые мы ей скажем. Таким образом, можно перемещаться по истории программы, вести паралелльно несколько версий программы, работать над ней сообща и многое другое.

Современная разработка немыслима без навыка GIT.

Чтобы сделать нашу обычную директорию "магической" GIT директорией (или репозиторием, как ее чаще называют), нам надо прописать команду git init. Она создаст в корне директории скрытую папку .git, в которой и будет хранить все изменения

Далее надо указать гиту, кто в нем будет работать, чтобы можно было понять, кто сделал те или иные изменения в коде.

Обычно, реальная разработка ведется более чем одним человеком, поэтому каждый из них обычно работает в своей версии проекта, так называемой ветке. Веткой по умолчанию, создаваемой при инициализации гита, является ветка master. Именно в ней всегда хранятся финальные, стабильные версии ПО.  

Чтобы код попал в мастер, разработчик должен сделать запрос на слияние (pull request) в эту ветку из своей ветки. Во время запроса на слияние, код проходит ревью другими разработчиками и, либо одобряется ими и попадает в мастер, либо отправляется на доработку.

Есть разные практики того, как правильно выстраивать процесс работы с гитом, но наибольшее распространение получили две из них: github flow и git flow.
